## Prepare Dataset

```python
twenties_male = df_le[(df_le["Year"]==2000) & (df_le["Sex"]==1)]
twenties_male = twenties_male[["Age", "People"]]

twenties_female = df_le[(df_le["Year"]==2000) & (df_le["Sex"]==2)]
twenties_female = twenties_female[["Age", "People"]]

nighties_male = df_le[(df_le["Year"]==1900) & (df_le["Sex"]==1)]
nighties_male = nighties_male[["Age", "People"]]

nighties_female = df_le[(df_le["Year"]==1900) & (df_le["Sex"]==2)]
nighties_female = nighties_female[["Age", "People"]]



### contructing twenties dataset
twenties_map = {}

vector = []

for k,v in twenties_male.items():
    for i in v:
        vector.append(i)

twenkeys = vector[:19]
twenvalues = vector[19:]
for i,j in zip(twenkeys, twenvalues):
    twenties_map[i] = j
#     print(i,j)
    
print()
vector = []

for k,v in twenties_female.items():
    for i in v:
        vector.append(i)

twenkeys = vector[:19]
twenvalues = vector[19:]
for i,j in zip(twenkeys, twenvalues):
    twenties_map[i] += j
#     print(i,j, twenties_map[i])


tdata = {'Age': [i for i in twenties_map.keys()],
        'People': [j for j in twenties_map.values()]} 
tframe = pd.DataFrame.from_dict(tdata)

### contructing nineteens dataset
nineteens_map = {}

vector = []

for k,v in nighties_male.items():
    for i in v:
        vector.append(i)

ninekeys = vector[:19]
ninevalues = vector[19:]
for i,j in zip(ninekeys, ninevalues):
    nineteens_map[i] = j
    print(i,j)
    
print()
vector = []

for k,v in nighties_female.items():
    for i in v:
        vector.append(i)

ninevalues = vector[19:]
for i in range(len(ninevalues)):
    ninevalues[i] *= -1

```

## The original plot

```python
#[Inser the code here]
#Example
# Set figure size for the notebook

fig, ax = plt.subplots()
(n, bins, patches) = ax.hist(x=twenkeys, weights=twenvalues,bins=37, edgecolor='white', label='population of 2000')
ax.set_xlabel('Age')
ax.set_ylabel('People')
ax.set_title('# equal-width bins = 20')

# fig1, ax1 = plt.subplots()
(n, bins, patches) = ax.hist(x=ninekeys, weights=ninevalues,bins=37, edgecolor='white', label='population of 1900')

plt.legend()
plt.show()
```
