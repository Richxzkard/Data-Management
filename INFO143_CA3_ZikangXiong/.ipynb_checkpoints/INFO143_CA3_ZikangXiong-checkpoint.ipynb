```python
#[Inser the code here]
#Example
# Dataset: Highest grossing movies for the weekend of December 22-24, 2017.

url_le = 'https://raw.githubusercontent.com/3eltran23/InfoViz/master/LifeExpectancyTheAmericas.csv'


df_le = pd.read_csv(url_le)
df_le = df_le.sort_values(by='2020') #sort by life expectancy
df_le.tail(5)
df_le.iloc[:,[0,-3, -4,-3,-2]].tail(10)
```

## The original plot

```python
#[Inser the code here]
#Example

fig, ax = plt.subplots(figsize=(7,10))

ax.scatter(df_le['2020'], df_le['Country Name']) #values, labels

ax.set_xlabel('Years')

# remove chart border
for spine in plt.gca().spines.values():
    spine.set_visible(False)


ax.grid()
```

# Label data Directly


```python
fig, ax = plt.subplots(figsize=(7,10))

ax.scatter(df_le['2020'], df_le['Country Name']) #values, labels
ax.set_xlabel('Years')

# for k, v in zip(df_le['2020'], df_le['Country Name']):
#     print(type)

locate = {}
indx = 1
for c in df_le['Country Name']:
    locate[c] = indx
    indx += 1

# remove chart border
for spine in plt.gca().spines.values():
    spine.set_visible(False)


for k, v in zip(df_le['2020'], locate.values()):
    plt.text(k+0.7,v-1.23, "{:.1f}".format(k))


ax.set_title('Life Expectancy of People of Countries at the America Continent in 2020', loc='left', fontsize=24,x=-0.29,  y=1.05)



ax.grid()
```

# Use sufficient contrast


```python
fig, ax = plt.subplots(figsize=(7,10))

# ax.scatter(df_le['2020'], df_le['Country Name']) #values, labels


gf = False
yf = False
of = False
rf = False

for i,j in zip(df_le['2020'], df_le['Country Name']):
    color = ""
    label = ""
    if i <= 82.5 and i >= 77.5:
        color = "green"
        label = "77.5 - 82.5"
    if i < 77.5 and i >= 72.5:
        color = "yellow"
        label = "72.5 - 77.5"
    if i < 72.5 and i >= 67.5:
        color = "orange"
        label = "67.5 - 72.5"
    if i < 67.5 and i >= 62.5:
        color = "red"
        label = "62.5 - 67.5"
        
    if color == "green":
        if not gf:
            ax.scatter(i, j, color = color, label = label)
            gf=True
            continue
    if color == "yellow":
        if not yf:
            ax.scatter(i, j, color = color, label = label)
            yf=True
            continue
    if color == "orange":
        if not of:
            ax.scatter(i, j, color = color, label = label)
            of=True
            continue
    if color == "red":
        if not rf:
            ax.scatter(i, j, color = color, label = label)
            rf=True
            continue
    ax.scatter(i, j, color = color)


ax.legend()
    


ax.set_xlabel('Years')

# for k, v in zip(df_le['2020'], df_le['Country Name']):
#     print(type)

locate = {}
indx = 1
for c in df_le['Country Name']:
    locate[c] = indx
    indx += 1

# remove chart border
for spine in plt.gca().spines.values():
    spine.set_visible(False)


for k, v in zip(df_le['2020'], locate.values()):
    plt.text(k+0.5,v-1.1, "{:.1f}".format(k))


ax.set_title('Life Expectancy of People of Countries at the America Continent in 2020', loc='left', fontsize=24,x=-0.29,  y=1.05)



ax.grid()
```
